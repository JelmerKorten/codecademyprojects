```python

# Write a function that returns a new list of updated damages where the recorded data is converted to float values and the missing data is retained as "Damages not recorded"

def convert_dmg(list):
    updated_list = []
    for i in list:
        if 'B' in i:
            strnumber = i[:-1]
            number = float(strnumber)
            newnumber = number * 1000000000
            updated_list.append(newnumber)
        elif 'M' in i:
            strnumber = i[:-1]
            number = float(strnumber)
            newnumber = number * 1000000
            updated_list.append(newnumber)
        else:
            updated_list.append(i)
    return updated_list


newdmgs = convert_dmg(damages)
print(newdmgs)
```

```python
def order_years(dict):
    years_dict = {}
    for all_names in dict.values():
        current_year = all_names.get("Year")
        current_cane = all_names
        if not current_year in years_dict:
            years_dict[current_year] = [current_cane]
        else:
            years_dict[current_year].append(current_cane)
    return(years_dict)
```

```python
def areacounter(dict):
    area_counts = {}
    for all_names in dict.values():
        areas = all_names.get("Areas Affected")
        for i in range(len(areas)):
            if not areas[i] in area_counts:
                area_counts[areas[i]] = 1
            else:
                area_counts[areas[i]] += 1
    return area_counts
```
